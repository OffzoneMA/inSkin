Display the values of { brands, code, categories, product_name, ingredients } for every document within the "products" collection of the "test" database.

```python
from pymongo import MongoClient

# Connect to the MongoDB server
client = MongoClient("mongodb://localhost:27017")

# Specify the database and collection
db = client["test"]
collection = db["products"]

# Specify the fields to project
fields_to_project = {
    "brands": 1,
    "code": 1,
    "categories": 1,
    "product_name": 1,
    "ingredients": 1,
    "tags": 1
}

# Helper function to recursively flatten dictionaries
def flatten_dict(d, parent_key='', sep='_'):
    items = []
    for k, v in d.items():
        new_key = parent_key + sep + k if parent_key else k
        if isinstance(v, dict):
            items.extend(flatten_dict(v, new_key, sep=sep).items())
        else:
            items.append((new_key, v))
    return dict(items)

# Iterate through the documents and print specified fields
for doc in collection.find({}, fields_to_project):
    print("brands:", doc.get("brands", ""))
    print("code:", doc.get("code", ""))
    print("categories:", doc.get("categories", ""))
    print("product_name:", doc.get("product_name", ""))
    
    ingredients = doc.get("ingredients")
    if ingredients:
        ingredient_list = []
        for ingredient in ingredients:
            if isinstance(ingredient, dict):
                flat_ingredient = flatten_dict(ingredient)
                ingredient_list.extend(flat_ingredient.values())
            else:
                ingredient_list.append(ingredient)
        
        print("ingredients:", ", ".join(map(str, ingredient_list)))
    
    tags = doc.get("tags")
    if tags:
        print("tags:", ", ".join(tags))

    print("-" * 40)

# Close the MongoDB connection
client.close()
```

Display the values of { brands, code, categories, product_name, ingredients } only for documents within the "products" collection of the "test" database, where the "categories" field contains the string "groceries" (case-insensitive).

```python
from pymongo import MongoClient
import re

# Connect to the MongoDB server
client = MongoClient("mongodb://localhost:27017")

# Specify the database and collection
db = client["test"]
collection = db["products"]

# Specify the fields to project
fields_to_project = {
    "brands": 1,
    "code": 1,
    "categories": 1,
    "product_name": 1,
    "ingredients": 1,
    "tags": 1
}

# Helper function to recursively flatten dictionaries
def flatten_dict(d, parent_key='', sep='_'):
    items = []
    for k, v in d.items():
        new_key = parent_key + sep + k if parent_key else k
        if isinstance(v, dict):
            items.extend(flatten_dict(v, new_key, sep=sep).items())
        else:
            items.append((new_key, v))
    return dict(items)

# Regular expression pattern to match "Groceries" in categories
groceries_pattern = re.compile(r'\bGroceries\b', re.IGNORECASE)

# Query to match documents with "Groceries" in the "categories" field
category_filter = {"categories": groceries_pattern}

# Iterate through the documents and print specified fields for matching documents
for doc in collection.find(category_filter, fields_to_project):
    print("brands:", doc.get("brands", ""))
    print("code:", doc.get("code", ""))
    print("categories:", doc.get("categories", ""))
    print("product_name:", doc.get("product_name", ""))
    
    ingredients = doc.get("ingredients")
    if ingredients:
        ingredient_list = []
        for ingredient in ingredients:
            if isinstance(ingredient, dict):
                flat_ingredient = flatten_dict(ingredient)
                ingredient_list.extend(flat_ingredient.values())
            else:
                ingredient_list.append(ingredient)
        
        print("ingredients:", ", ".join(map(str, ingredient_list)))
    
    tags = doc.get("tags")
    if tags:
        print("tags:", ", ".join(tags))

    print("-" * 40)

# Close the MongoDB connection
client.close()
```

Clean collection products inside off db on mongo client to export products containing "Open Beauty Facts" in their categories field to matching_products collection in the same db

```python
from pymongo import MongoClient
import re

# Connect to the MongoDB server
client = MongoClient("mongodb://localhost:27017")

# Specify the database and source collection
db = client["off"]
source_collection = db["products"]

# Specify the target collection for storing matching documents
target_collection = db["matching_products"]

# Regular expression pattern to match "Groceries" in categories
openbeautyfacts_pattern = re.compile(r'\bOpen Beauty Facts\b', re.IGNORECASE)

# Query to match documents with "Groceries" in the "categories" field
category_filter = {"categories": openbeautyfacts_pattern}

# Iterate through the documents and push matching documents to the target collection
for doc in source_collection.find(category_filter):
    print(f"Processing document _id: {doc['_id']}")
    target_collection.insert_one(doc)

# Close the MongoDB connection
client.close()
```

Retrieve 20 random documents from mongo client "openbeautyfacts_products" in "off" db collection and store them in "products" collection inside "tempoff" db.

```python
import random
from pymongo import MongoClient

# MongoDB connection settings
db_name = "off"
collection_name = "openbeautyfacts_products"
new_db_name = "tempoff"
new_collection_name = "products"

# Connect to the MongoDB server
client = MongoClient("mongodb://localhost:27017/")
db = client[db_name]
new_db = client[new_db_name]

# Get the total number of documents in the source collection
total_documents = db[collection_name].count_documents({})

# Generate a list of random indices to select documents
random_indices = random.sample(range(total_documents), 20)

# Retrieve and insert random documents into the new collection
for index in random_indices:
    random_document = db[collection_name].find().limit(1).skip(index).next()
    new_db[new_collection_name].insert_one(random_document)

# Close the MongoDB connection
client.close()

print("Documents migrated successfully.")
```
